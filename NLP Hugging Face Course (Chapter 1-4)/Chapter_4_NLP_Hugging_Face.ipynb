# **Using pretrained models**

**Model Hub** membuat pemilihan model yang sesuai menjadi sederhana, sehingga penggunaannya dalam pustaka downstream mana pun dapat dilakukan hanya dengan beberapa baris kode. Mari kita lihat bagaimana cara sebenarnya menggunakan salah satu model ini, dan bagaimana memberikan kontribusi kembali ke komunitas.

Misalnya, kita mencari model berbasis bahasa Prancis yang dapat melakukan **mask filling**.  
Kita memilih checkpoint **camembert-base** untuk mencobanya. Identifier **camembert-base** adalah semua yang kita butuhkan untuk mulai menggunakannya! Seperti yang telah Anda lihat di bab-bab sebelumnya, kita dapat menginstansikannya menggunakan fungsi `pipeline()`:

```python
from transformers import pipeline

camembert_fill_mask = pipeline("fill-mask", model="camembert-base")
results = camembert_fill_mask("Le camembert est <mask> :)")
```

Anda juga dapat menginstansiasi checkpoint menggunakan arsitektur model secara langsung:

```python
from transformers import CamembertTokenizer, CamembertForMaskedLM

tokenizer = CamembertTokenizer.from_pretrained("camembert-base")
model = CamembertForMaskedLM.from_pretrained("camembert-base")
```

Namun, kami merekomendasikan menggunakan kelas **Auto*** sebagai gantinya, karena kelas ini secara desain bersifat **architecture-agnostic** (tidak bergantung pada arsitektur tertentu). Sementara contoh kode sebelumnya membatasi pengguna pada checkpoint yang dapat dimuat dalam arsitektur **CamemBERT**, penggunaan kelas **Auto*** mempermudah pergantian checkpoint:

```python
from transformers import AutoTokenizer, AutoModelForMaskedLM

tokenizer = AutoTokenizer.from_pretrained("camembert-base")
model = AutoModelForMaskedLM.from_pretrained("camembert-base")
```
